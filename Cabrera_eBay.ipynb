## Clean Data
- Brand column: make all lowercase, and standardize the brands
- condition

```python
# view brand column
print(merged_df['Brand'].unique())

# make everything lowercase
merged_df['Brand'] = merged_df['Brand'].str.lower()
```

```python
 # view row with btand 'hp dell lenovo select from above'
 merged_df[merged_df['Brand'] == 'hp dell lenovo select from above']

 # remove this row
 merged_df = merged_df[merged_df['Brand'] != 'hp dell lenovo select from above']
```

```python
# replace dell inc and dell incpirior with dell
merged_df['Brand'] = merged_df['Brand'].replace(['dell inc.', 'dell inspiron', 'dell/hp/lenovo/stone'], 'dell')
merged_df['Brand'] = merged_df['Brand'].replace(['lenovo t440'], 'lenovo')
```

```python
# condition column
print(merged_df['Condition'].unique())
# make lowercase
merged_df['Condition'] = merged_df['Condition'].str.lower()
```

```python
def clean_condition(condition):
    if pd.isna(condition):
        return 'unknown'

    # Use more specific checks for "new" to avoid miscategorizing
    condition_lower = condition.lower()

    if condition_lower.startswith('new'):
        return 'new'
    elif 'excellent' in condition_lower:
        return 'excellent'
    elif 'refurbished' in condition_lower:
        if 'very good' in condition_lower:
            return 'refurbished - very good'
        elif 'good' in condition_lower:
            return 'refurbished - good'
        elif 'excellent' in condition_lower:
            return 'refurbished - excellent'
        else:
            return 'refurbished'
    elif 'used' in condition_lower:
        return 'used'
    elif 'seller refurbished' in condition_lower:
        return 'seller refurbished'
    elif 'certified' in condition_lower:
        return 'certified refurbished'
    else:
        return 'other'

# Apply the function to clean condition column
merged_df['Condition'] = merged_df['Condition'].apply(clean_condition)
```

```python
merged_df.head()
```

```python
def clean_price(price):
    # Check for ranges and take the average if it's a range
    if 'to' in price:
        prices = price.replace('$', '').replace(',', '').split(' to ')
        return (float(prices[0]) + float(prices[1])) / 2
    else:
        # Remove dollar sign, commas and convert to float
        return float(price.replace('$', '').replace(',', '').strip())

# apply to price
merged_df['Price'] = merged_df['Price'].apply(clean_price)

#view
merged_df.head()
```

```python
# Remove duplicates based on 'Item Number'
merged_df = merged_df.drop_duplicates(subset=['Item Number'], keep='first')

# View the updated DataFrame
print(merged_df.shape)
```
